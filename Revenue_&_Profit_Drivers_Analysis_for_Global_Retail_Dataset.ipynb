```python
import pandas as pd
```

```python
df = pd.read_csv('Global Superstore.csv')
revenue = df['Sales'].sum()
profit = df['Profit'].sum()
orders = df['Order ID'].nunique()
customers = df['Customer ID'].nunique()

avg_order_value = revenue / orders
avg_order_profit = profit / orders

print("Revenue:", round(revenue, 2))
print("Profit:", round(profit, 2))
print("Orders:", orders)
print("Customers:", customers)
print("Average Order Value:", round(avg_order_value, 2))
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
category_orders = df.groupby('Category')['Order ID'].nunique().sort_values(ascending=False)

print("Sales by Category:")
print(category_sales)
print("\n Profit by Category:")
print(category_profit)
print("\n Orders by Category:")
print(category_orders)
```

```python
margin = profit/revenue
print(round(margin * 100,1))
# Маржинальность составляет 11.6%, бизнес работает с положительной прибылью, главный показатель выручки - Technology
```

```python
category_margin = (df.groupby("Category").agg({"Sales": "sum","Profit": "sum"}))

category_margin["Margin %"] = (category_margin["Profit"] /category_margin["Sales"] * 100)

print(category_margin)
```

```python
country_sales = (df.groupby("Country").agg({"Sales": "sum","Profit": "sum"}).sort_values("Sales", ascending=False))

print(country_sales.head(10))
```

```python
country_profit = (df.groupby("Country")["Profit"].sum().sort_values(ascending=False))

print(country_profit.head(10))
```

```python
technology_discount = (df[df["Category"] == "Technology"]["Discount"].mean())
ofice_discount = (df[df["Category"] == "Office Supplies"]["Discount"].mean())
furniture_discount = (df[df["Category"] == "Furniture"]["Discount"].mean())

print(ofice_discount)
print(furniture_discount)

print(technology_discount)
```

```python
technology_shipping = (df[df["Category"] == "Technology"]["Ship Mode"].value_counts())
ofice_shipping = (df[df["Category"] == "Office Supplies"]["Ship Mode"].value_counts())
furniture_shipping = (df[df["Category"] == "Furniture"]["Ship Mode"].value_counts())
print(technology_shipping)
print(ofice_shipping)
print(furniture_shipping)
```

```python
technology_shipping_cost = (df[df["Category"] == "Technology"]["Shipping Cost"].mean())
ofice_shipping_cost = (df[df["Category"] == "Office Supplies"]["Shipping Cost"].mean())
furniture_shipping_cost = (df[df["Category"] == "Furniture"]["Shipping Cost"].mean())

print('technology_shipping_cost',technology_shipping_cost)
print('\n ofice_shipping_cost',ofice_shipping_cost)
print('\n furniture_shipping_cost',furniture_shipping_cost)
```
